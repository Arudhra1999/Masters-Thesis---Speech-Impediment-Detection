```python
import os
import pandas as pd
import whisper
import librosa
import numpy as np
import torch
```

```python
data = pd.read_csv('D:\\Time Series EDA\\SEP-28k_labels - 1.csv')
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device : {device}")

model = whisper.load_model('tiny').to(device)
```

```python
transcripts = []
labels = []

base_audio_dir = 'Clipped_Audio'
```

```python
for index, row in data.iterrows():
    folder_name = row['Show']
    audio_id = row['EpId']
    clip_id = row['ClipId']
    label = row['Original Label']
   
    audio_path = os.path.join(base_audio_dir, folder_name, f"{audio_id}", f"{folder_name}_{audio_id}_{clip_id}.wav")

    if os.path.exists(audio_path):
        result = model.transcribe(audio_path)
        transcript = result['text']
        transcripts.append(transcript)
        labels.append(label)
        print(f"Successfully Transcribed : {audio_path}")
    else:
        print(f"File not found: {audio_path}")
```

```python
df_transcripts = pd.DataFrame({'Transcript': transcripts, 'Label': labels})
df_transcripts.to_csv('transcripts_with_labels_1.csv', index=False)
```
